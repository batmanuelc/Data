```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("population_by_country_2020.csv")
df

eastAsia = df[df["Country"].isin(["China", "South Korea", "North Korea", "Macau", "Hong Kong", "Tibet", "Taiwan", "Japan", "Mongolia"])]

country = eastAsia["Country"]
population = eastAsia["Population (2020)"]

fig = plt.figure(figsize = (16,9))

plt.barh(country,population)
plt.title("2020 Population by Country (East Asia)")

plt.show
```
