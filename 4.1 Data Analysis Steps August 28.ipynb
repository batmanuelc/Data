```python
#Import packages and libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the csv file into a DataFrame
train = pd.read_csv('train.csv')
#See the first 5 rows
train.head(5)
```

```python
#Checking if there is a null values.
train.isnull().sum()
```

```python
#Checking it the postal code further.
train[train['Postal Code'].isna()]
```

```python
#Looks like Burlington have Null postal codes
train[train['City'] == 'Burlington']
```

```python
#Filling in null values with non-null values
train['Postal Code'] = train['Postal Code'].fillna('05401')
```

```python
train[train['City'] == 'Burlington']
```

```python
train.info()
train.isnull().sum()
```

```python
train.head()
```

```python
#Deleting a column
del train['Row ID']
train.head(2)
```

```python
#Checking and Converting Data types
train.info()
```

```python
#Converting object to datetime
#1st Step is
#train['Ship Date'] = pd.to_datetime(train['Ship Date'])
#2nd Step is 
#train['Ship Date'] = pd.to_datetime(train['Ship Date'], format = '%d%m%Y')


train['Ship Date'] = pd.to_datetime(train['Ship Date'], format = '%d%m%Y')
train['Order Date'] = pd.to_datetime(train['Order Date'], format = '%d%m%Y')

#Converting float to object or string
train['Postal Code'] = train['Postal Code'].astype(str)

train.info()
```

```python
train.head(2)
```

## Solving the Questions Above

#### 1. What is the yearly sales trend?

```python
#Create order year column
train['Order Year'] = train['Order Date'].dt.year

train.info()
train.head(2)
```

```python
#Convert int into object
train['Order Year'] = train['Order Year'].astype(str)
train.info()
```

```python
#Sum of Sales by year
#Final Answer

yearSales = train.groupby('Order Year', as_index = False).agg({'Sales':'sum'})
yearSales
```

```python
sns.lineplot(x='Order Year', y='Sales', data=yearSales, marker='o')
plt.xlabel('')
plt.show
```

#### 2. Which product sold the most in 2018?


```python
sales2018 = train[train['Order Year'] == '2018']
sales2018.head(2)
```

```python
productSold = sales2018.groupby('Product Name', as_index = False).agg({'Sales':'sum'})
output = sales2018.sort_values(by = 'Sales', ascending = False)
output.head(10)
```

```python
p_top = output[(output['Sales']>=4416.174)]
sns.barplot(y="Product Name", x="Sales", data=p_top, ci=False, orient='h', color='grey')
plt.title('top products in 2018')
plt.ylabel('')
plt.show()
```

#### 3. Among the three product categories, which one had the lowest sale in 2018?


```python
lowestSale = sales2018.groupby('Category', as_index = False).agg({'Sales':'sum'})
lowestSale
```

```python
sns.barplot(x='Category', y='Sales', data=lowestSale, palette = 'Blues')
plt.title('Sales by Product category')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

#### 4. What is the yearly sales trend of the three customer groups?



```python
consumerSegment = train[train['Segment'] == 'Consumer']
consumer = consumerSegment.groupby('Order Year', as_index = False).agg({'Sales':'sum'})
consumer
```

```python
homeSegment = train[train['Segment'] == 'Home Office']
home = homeSegment.groupby('Order Year', as_index = False).agg({'Sales':'sum'})
home
```

```python
corpSegment = train[train['Segment'] == 'Corporate']
corp = corpSegment.groupby('Order Year', as_index = False).agg({'Sales':'sum'})
corp
```

```python
sns.lineplot(x='Order Year', y='Sales', data=consumer, marker='o')
sns.lineplot(x='Order Year', y='Sales', data=home, marker='o')
sns.lineplot(x='Order Year', y='Sales', data=corp, marker='o')

plt.legend(['consumer', 'home', 'corporate'], fontsize=8)
plt.show
```
